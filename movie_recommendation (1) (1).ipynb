```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
a=pd.read_csv("/content/movies.csv")
a
```

```python
selected_features=[['genres', 'keywords', 'overview', 'tagline',
    'cast', 'crew', 'director',
    'original_language', 'production_companies']]
print(selected_features)
```

```python
for features in selected_features:
  a[features]=a[features].fillna('')
```

```python
a.isna().sum()
```

```python
combined_features=a['genres'] + " "+a['keywords'] + " " +a['overview'] + " " +a['tagline'] + " " +a['cast'] + " " +a['crew'] + " " +a['director'] + " " +a  ['original_language'] + " " +a['production_companies']
print(combined_features)
```

```python
Vectorizer=TfidfVectorizer()
feature_Vectors=Vectorizer.fit_transform(combined_features)
print(feature_Vectors)
```

```python
similarity=cosine_similarity(feature_Vectors)
print(similarity)
```

```python
movie_name=input("enter your favourite movie name: ")
```

```python
list_of_all_titles=a['title'].tolist()
print(list_of_all_titles)
```

```python
import difflib

find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles)
close_match=find_close_match[0]
print(close_match)
```

```python
close_match=find_close_match[0]
print(close_match)
```

```python
index_of_movie=a[a.title==close_match]['index'].values[0]
print(index_of_movie)
```

```python
similarity_score=list(enumerate(similarity[index_of_movie]))
print(similarity_score)
```

```python
sorted_similar_movies=sorted(similarity_score,key=lambda i:i[1], reverse=True)
print(sorted_similar_movies)
```

```python
print("movies suggested for you : \i")

i=1

for movie in sorted_similar_movies:
  index=movie[0]
  title_from_index=a[a.index==index]['title'].values[0]
  if(i<30):
    print(i, '.',title_from_index)
    i=i=1
```
